```python
"""
Data for analysis

Import, analyse and transform the provided data.
"""

import pandas as pd 

df = pd.read_excel('Teste BI base.xlsx')

df.info()
display(df.head())
```

```python
"""
Transform the columns
"""

# Convert 'velocidade' to float
df['velocidade'] = df['velocidade'].astype(str).str.replace(',', '.').astype(float)

# Convert 'data' and 'dataserver' to datetime
df['data'] = pd.to_datetime(df['data'])
df['dataserver'] = pd.to_datetime(df['dataserver'])

```

```python
df.info()
display(df.head())
```

```python
# Save the cleaned DataFrame to an Excel file
df.to_excel('_clean_Teste BI base.xlsx', index=False)
```
